#Q1.What is hypothesis testing in statistics?

 - Hypothesis testing in statistics is a formal procedure used to make decisions about a population based on sample data. It involves formulating a null hypothesis and an alternative hypothesis, collecting data, and then using statistical methods to determine whether there is enough evidence to reject the null hypothesis.


 #Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

  - The null hypothesis (H0) is a statement of no effect or no difference, essentially proposing that any observed difference is due to random chance. The alternative hypothesis (H1 or Ha) is a statement that contradicts the null hypothesis, suggesting that there is a significant effect or difference. For example, if testing a new drug, the null hypothesis might be that the drug has no effect, while the alternative hypothesis would be that it does have an effect.

#Q3.What is the significance level in hypothesis testing, and why is it important?

 - The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true. It represents the maximum probability of making a Type I error that a researcher is willing to accept. It is important because it sets the standard for how much evidence is needed to reject the null hypothesis. Common significance levels are 0.05 or 0.01.


#Q4.What does a P-value represent in hypothesis testing?

 - A P-value represents the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming that the null hypothesis is true.


#Q5.How do you interpret the P-value in hypothesis testing?


- In hypothesis testing, the P-value is compared to the significance level (α).

 - If the P-value is less than or equal to α (P≤α), you reject the null hypothesis. This suggests that the observed data is unlikely to have occurred if the null hypothesis were true, providing evidence in favor of the alternative hypothesis.

 - If the P-value is greater than α (P>α), you fail to reject the null hypothesis. This means there isn't enough statistical evidence to conclude that the alternative hypothesis is true.


#Q6.What are Type 1 and Type 2 errors in hypothesis testing?

 - Type 1 error (False Positive): Occurs when you reject a true null hypothesis. The probability of a Type 1 error is denoted by α (the significance level).

 - Type 2 error (False Negative): Occurs when you fail to reject a false null hypothesis. The probability of a Type 2 error is denoted by β.


#Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

- The difference lies in the alternative hypothesis and the direction of the effect being tested.
 -  One-tailed test: The alternative hypothesis specifies a direction (e.g., the mean is greater than a certain value, or less than a certain value). The critical region for rejection is entirely in one tail of the sampling distribution.
 - Two-tailed test: The alternative hypothesis states that there is a difference, but doesn't specify the direction (e.g., the mean is not equal to a certain value). The critical region for rejection is split between both tails of the sampling distribution.


#Q8.What is the Z-test, and when is it used in hypothesis testing?

 - The Z-test is a statistical hypothesis test that is used when the population standard deviation is known, or when the sample size is large (typically n≥30), allowing the sample standard deviation to be a good estimate of the population standard deviation. It is used to test hypotheses about population means.


#Q9.How do you calculate the Z-score, and what does it represent in hypothesis testing

- The Z-score is calculated as:Z=(X−μ)/(σ/√n)

where:

- X is the sample mean
- μ is the population mean
- σ is the population standard deviation
- n is the sample size

In hypothesis testing, the Z-score represents how many standard deviations a data point or sample mean is away from the population mean. It allows you to determine the probability of observing a particular sample mean if the null hypothesis were true.


#Q10.What is the T-distribution, and when should it be used instead of the normal distribution?

- The T-distribution (Student's t-distribution) is a probability distribution that is similar to the normal distribution but has fatter tails. It should be used instead of the normal distribution when the population standard deviation is unknown and the sample size is small (typically n<30).
As the sample size increases, the T-distribution approaches the normal distribution.


#Q11.What is the difference between a Z-test and a T-test?

- The primary difference between a Z-test and a T-test lies in the knowledge of the population standard deviation and the sample size.

   - Z-test: Used when the population standard deviation (σ) is known, or when the sample size (n) is large (n≥30).
   - T-test: Used when the population standard deviation (σ) is unknown and the sample size (n) is small (n<30). The T-test uses the sample standard deviation as an estimate for the population standard deviation.


#Q12.What is the T-test, and how is it used in hypothesis testing?

- The T-test is a statistical hypothesis test used to compare means when the population standard deviation is unknown and the sample size is small. It is commonly used to:

  - Compare a sample mean to a known population mean (one-sample t-test).
  - Compare the means of two independent groups (independent samples t-test).
  - Compare the means of two related or paired groups (paired samples t-test). The T-test calculates a t-statistic, which is then compared to a critical value from the t-distribution to make a decision about the null hypothesis.


#Q13.What is the relationship between Z-test and T-test in hypothesis testing?

- Both the Z-test and T-test are inferential statistical tests used for hypothesis testing concerning population means. The T-test is essentially an adaptation of the Z-test for situations where the population standard deviation is unknown and must be estimated from the sample. As the sample size increases, the T-distribution approximates the normal distribution, and thus, for large sample sizes, the results of a T-test will approach those of a Z-test.


#Q14.What is a confidence interval, and how is it used to interpret statistical results?

 - A confidence interval is a range of values that is likely to contain the true population parameter (e.g., population mean, population proportion) with a certain level of confidence. It is used to interpret statistical results by providing a range of plausible values for the true population parameter rather than just a single point estimate. For example, a 95% confidence interval for a mean suggests that if you were to repeat the sampling process many times, 95% of the calculated intervals would contain the true population mean.



#15.What is the margin of error, and how does it affect the confidence interval?

 - The margin of error is the half-width of the confidence interval. It quantifies the amount of random sampling error in a survey's results. A larger margin of error leads to a wider confidence interval, indicating less precision in estimating the population parameter. Conversely, a smaller margin of error results in a narrower confidence interval, suggesting greater precision


#Q16.How is Bayes' Theorem used in statistics, and what is its significance

 - Bayes' Theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event[cite: 15]. It is used in statistics for Bayesian inference, which updates the probability of a hypothesis as more evidence or information becomes available.

The formula is:  = P(A∣B)=P(B∣A)∗P(A)/P(B)

 where:

     -P(A∣B) is the posterior probability of event A given event B
     -P(B∣A) is the likelihood of event B given event A
     -P(A) is the prior probability of event A
     -P(B) is the marginal probability of event B

Its significance lies in its ability to combine prior beliefs with new data to produce updated probabilities, making it powerful for decision-making under uncertainty and for building sophisticated statistical models.


#Q17.What is the Chi-square distribution, and when is it used?

- The Chi-square (χ2) distribution is a non-negative, asymmetric distribution used primarily in hypothesis testing. It is used when:

  - Testing for independence between two categorical variables (Chi-square test of independence).

  - Testing whether observed frequencies differ significantly from expected frequencies (Chi-square goodness-of-fit test).

  - Estimating population variance or standard deviation.

#Q19.What is the F-distribution, and when is it used in hypothesis testing?

 The F-distribution is a continuous probability distribution that arises in the testing of hypotheses concerning the equality of variances from two normally distributed populations, or in the context of analysis of variance (ANOVA). It is used in hypothesis testing when:

- Comparing the variances of two populations (F-test for equality of variances).
- Performing an ANOVA test to compare the means of three or more groups.



#Q20.What is an ANOVA test, and what are its assumptions?

- ANOVA stands for Analysis of Variance. It is a statistical test used to compare the means of three or more independent groups to determine if at least one group mean is significantly different from the others[cite: 19]. Instead of performing multiple t-tests (which would increase the chance of Type I error), ANOVA examines the variance between groups relative to the variance within groups.

The assumptions of an ANOVA test are:

 - Independence of observations: The observations within each group, and between groups, must be independent.
 -  Normality: The data within each group should be approximately normally distributed.

#Q21.What are the different types of ANOVA tests?

 The different types of ANOVA tests include:

 - One-Way ANOVA: Used to compare the means of three or more independent groups based on one categorical independent variable.
 - Two-Way ANOVA: Used to compare the means of groups that are classified based on two independent categorical variables, and to examine the interaction between these two variables.
 - MANOVA (Multivariate Analysis of Variance): Used when there are multiple dependent variables.
 - Repeated Measures ANOVA: Used when the same subjects are measured under different conditions or at different time points.


#Q22.What is the F-test, and how does it relate to hypothesis testing?

- The F-test is a statistical test that uses the F-distribution to compare variances or to compare means across multiple groups. In the context of hypothesis testing, the F-test is most commonly used in ANOVA to determine if the variability between group means is significantly larger than the variability within the groups. If the F-statistic is large enough (and the corresponding P-value is small), it suggests that there are significant differences between the group means, leading to the rejection of the null hypothesis that all group means are equal.

```python
#Q9.Write a Python function to calculate Z-scores from a dataset and plot them.
'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates the Z-scores for each data point in a dataset
    and plots a histogram of the Z-scores.

    Args:
        data (list or np.array): The input dataset.

    Returns:
        np.array: An array of Z-scores.
    """
    if not isinstance(data, (list, np.ndarray)):
        raise TypeError("Input data must be a list or NumPy array.")
    if len(data) < 2:
        raise ValueError("Dataset must contain at least two data points to calculate Z-scores.")

    data_np = np.array(data)
    mean = np.mean(data_np)
    std_dev = np.std(data_np)

    if std_dev == 0:
        print("Warning: Standard deviation is zero. All Z-scores will be 0.")
        z_scores = np.zeros_like(data_np, dtype=float)
    else:
        z_scores = (data_np - mean) / std_dev

    # Plotting the Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='purple', edgecolor='black')
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency/Probability Density')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add theoretical normal curve for comparison (Z-scores should be standard normal)
    from scipy.stats import norm
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, 0, 1) # Standard normal distribution (mean=0, std=1)
    plt.plot(x, p, 'r', linewidth=2, label='Standard Normal PDF')
    plt.legend()
    plt.show()

    return z_scores

# --- Example Usage ---
# Example 1: Data from a normal distribution
data_normal = np.random.normal(loc=50, scale=10, size=1000)
print("Z-scores for normally distributed data:")
z_scores_normal = calculate_and_plot_z_scores(data_normal)
print(f"First 5 Z-scores: {z_scores_normal[:5]}")
print(f"Mean of Z-scores: {np.mean(z_scores_normal):.4f}")
print(f"Standard deviation of Z-scores: {np.std(z_scores_normal):.4f}\n")


# Example 2: Skewed data
data_skewed = np.random.exponential(scale=2, size=1000)
print("Z-scores for exponentially distributed data:")
z_scores_skewed = calculate_and_plot_z_scores(data_skewed)
print(f"First 5 Z-scores: {z_scores_skewed[:5]}")
print(f"Mean of Z-scores: {np.mean(z_scores_skewed):.4f}")
print(f"Standard deviation of Z-scores: {np.std(z_scores_skewed):.4f}")
'''
```

```python
#Q10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
'''
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the non-normal distribution (e.g., Exponential distribution)
# The CLT states that the distribution of sample means will be approximately normal,
# regardless of the original distribution, given a large enough sample size for each sample.
lambda_param = 0.5 # Rate parameter for exponential distribution (mean = 1/lambda)

# Simulation parameters
sample_size = 30       # Size of each sample (should be >= 30 for CLT to hold well)
num_samples = 1000     # Number of samples to draw

# Generate sample means
sample_means = []
for _ in range(num_samples):
    # Draw a sample from the exponential distribution
    exponential_sample = np.random.exponential(scale=1/lambda_param, size=sample_size)
    # Calculate the mean of the sample
    sample_means.append(np.mean(exponential_sample))

# Convert to NumPy array for easier calculations
sample_means = np.array(sample_means)

# Plot the histogram of sample means
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='teal', edgecolor='black')
plt.title(f'Distribution of Sample Means (n={sample_size}) from Exponential Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency/Probability Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Overlay a normal distribution with the expected mean and standard deviation
# Expected mean of sample means = mean of original distribution (1/lambda_param)
# Expected standard deviation of sample means = (std dev of original distribution) / sqrt(sample_size)
# For exponential distribution, std dev = mean = 1/lambda_param
expected_mean_of_sample_means = 1 / lambda_param
expected_std_dev_of_sample_means = (1 / lambda_param) / np.sqrt(sample_size)

from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, expected_mean_of_sample_means, expected_std_dev_of_sample_means)
plt.plot(x, p, 'r', linewidth=2, label=f'Theoretical Normal PDF (μ={expected_mean_of_sample_means:.2f}, σ={expected_std_dev_of_sample_means:.2f})')
plt.legend()
plt.show()

print(f"Original distribution (Exponential) mean: {1/lambda_param:.2f}")
print(f"Mean of sample means: {np.mean(sample_means):.2f}")
print(f"Standard deviation of sample means: {np.std(sample_means):.2f}")
print(f"Theoretical standard error (std dev / sqrt(n)): {expected_std_dev_of_sample_means:.2f}")
'''
```

```python
#Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def calculate_z_score(data_point, dataset_mean, dataset_std_dev):
    """
    Calculates the Z-score for a single data point.

    Args:
        data_point (float): The individual data point for which to calculate the Z-score.
        dataset_mean (float): The mean of the dataset.
        dataset_std_dev (float): The standard deviation of the dataset.

    Returns:
        float: The calculated Z-score.
    """
    # Handle the case where standard deviation is zero to prevent division by zero.
    # In such a case, all data points are identical, and their Z-scores would be 0.
    if dataset_std_dev == 0:
        return 0.0
    return (data_point - dataset_mean) / dataset_std_dev

# --- Example Usage ---
# Sample dataset: A set of numerical values
data = [65, 70, 72, 75, 80, 82, 85, 88, 90, 95]
data_np = np.array(data) # Convert the list to a NumPy array for easier calculations

# Calculate the mean and standard deviation of the dataset
dataset_mean = np.mean(data_np)
dataset_std_dev = np.std(data_np) # Population standard deviation by default in numpy, or use ddof=1 for sample std dev

print(f"Dataset: {data_np}")
print(f"Dataset Mean: {dataset_mean:.2f}")
print(f"Dataset Standard Deviation: {dataset_std_dev:.2f}\n")

# Choose a specific data point from the dataset to calculate its Z-score
given_data_point = 85
z_score = calculate_z_score(given_data_point, dataset_mean, dataset_std_dev)
print(f"Z-score for data point {given_data_point}: {z_score:.4f}")

# --- Compare the Z-score to a standard normal distribution ---
# The Z-score tells us how many standard deviations an individual data point is
# away from the mean of its dataset. We can then use the standard normal distribution
# (which has a mean of 0 and a standard deviation of 1) to understand the probability
# associated with that Z-score.

# Probability of observing a value less than or equal to the calculated Z-score (Cumulative Distribution Function - CDF)
p_value_less_than = norm.cdf(z_score)
print(f"Probability (P(Z <= {z_score:.2f})): {p_value_less_than:.4f}")

# Probability of observing a value greater than or equal to the calculated Z-score (1 - CDF)
p_value_greater_than = 1 - norm.cdf(z_score)
print(f"Probability (P(Z >= {z_score:.2f})): {p_value_greater_than:.4f}")

# --- Visualize the Z-score on a standard normal distribution plot ---
plt.figure(figsize=(10, 6)) # Create a new figure for the plot

# Generate x values for plotting the standard normal distribution (typically from -4 to +4 std dev)
x = np.linspace(-4, 4, 1000)

# Plot the Probability Density Function (PDF) of the standard normal distribution
plt.plot(x, norm.pdf(x, 0, 1), color='blue', label='Standard Normal PDF (μ=0, σ=1)')

# Add a vertical dashed line at the calculated Z-score to highlight its position
plt.axvline(z_score, color='red', linestyle='--', label=f'Data Point Z-score: {z_score:.2f}')

# Shade the area under the curve to the left of the Z-score, representing P(Z <= z_score)
plt.fill_between(x, 0, norm.pdf(x, 0, 1), where=(x <= z_score), color='skyblue', alpha=0.3, label=f'P(Z <= {z_score:.2f})')

plt.title('Z-score on Standard Normal Distribution') # Set the title of the plot
plt.xlabel('Z-score') # Label for the x-axis
plt.ylabel('Probability Density') # Label for the y-axis
plt.grid(True, linestyle='--', alpha=0.7) # Add a grid for better readability
plt.legend() # Display the legend for labels
plt.show() # Show the plot

# --- Another example with a data point further from the mean ---
given_data_point_extreme = 100
z_score_extreme = calculate_z_score(given_data_point_extreme, dataset_mean, dataset_std_dev)
print(f"\nZ-score for extreme data point {given_data_point_extreme}: {z_score_extreme:.4f}")
print(f"Probability (P(Z >= {z_score_extreme:.2f})): {1 - norm.cdf(z_score_extreme):.4f}")
```

```python
#Q15.Implement hypothesis testing using Z-statistics for a sample dataset.
'''
import numpy as np
from scipy.stats import norm

def z_test_one_sample(sample_mean, population_mean_hypothesized, population_std_dev, sample_size, alpha=0.05, test_type='two-tailed'):
    """
    Performs a one-sample Z-test.

    Args:
        sample_mean (float): The mean of the sample.
        population_mean_hypothesized (float): The hypothesized population mean under the null hypothesis.
        population_std_dev (float): The known population standard deviation.
        sample_size (int): The size of the sample.
        alpha (float): The significance level (default is 0.05).
        test_type (str): Type of test ('two-tailed', 'left-tailed', 'right-tailed').

    Returns:
        tuple: A tuple containing the Z-statistic, P-value, and the decision.
    """
    if population_std_dev <= 0:
        raise ValueError("Population standard deviation must be positive.")
    if sample_size <= 0:
        raise ValueError("Sample size must be positive.")

    # Calculate the standard error of the mean
    standard_error = population_std_dev / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean_hypothesized) / standard_error

    # Calculate the P-value based on the test type
    if test_type == 'two-tailed':
        p_value = 2 * norm.cdf(-abs(z_statistic))
    elif test_type == 'left-tailed':
        p_value = norm.cdf(z_statistic)
    elif test_type == 'right-tailed':
        p_value = 1 - norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid test_type. Choose from 'two-tailed', 'left-tailed', 'right-tailed'.")

    # Make the decision
    decision = "Fail to reject the null hypothesis"
    if p_value < alpha:
        decision = "Reject the null hypothesis"

    return z_statistic, p_value, decision

# --- Example Usage ---
# Scenario: A school claims that the average IQ of its students is 110.
# We take a sample of 40 students and find their average IQ is 115.
# Assume the population standard deviation of IQ is known to be 15.
# We want to test if the school's claim is false (i.e., the mean IQ is not 110).

sample_mean_iq = 115
hypothesized_population_mean_iq = 110
known_population_std_dev_iq = 15
sample_size_iq = 40
significance_level = 0.05

print("--- One-Sample Z-Test Example ---")
print(f"Sample Mean: {sample_mean_iq}")
print(f"Hypothesized Population Mean (H0): {hypothesized_population_mean_iq}")
print(f"Known Population Standard Deviation: {known_population_std_dev_iq}")
print(f"Sample Size: {sample_size_iq}")
print(f"Significance Level (alpha): {significance_level}\n")

# Perform a two-tailed test
z_stat, p_val, decision_result = z_test_one_sample(
    sample_mean=sample_mean_iq,
    population_mean_hypothesized=hypothesized_population_mean_iq,
    population_std_dev=known_population_std_dev_iq,
    sample_size=sample_size_iq,
    alpha=significance_level,
    test_type='two-tailed'
)

print(f"Calculated Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Decision: {decision_result}")

if decision_result == "Reject the null hypothesis":
    print(f"Conclusion: There is significant evidence to suggest that the true average IQ is not {hypothesized_population_mean_iq}.")
else:
    print(f"Conclusion: There is not enough significant evidence to reject the claim that the true average IQ is {hypothesized_population_mean_iq}.")

print("\n--- Right-Tailed Test Example (Is the mean IQ GREATER than 110?) ---")
z_stat_right, p_val_right, decision_result_right = z_test_one_sample(
    sample_mean=sample_mean_iq,
    population_mean_hypothesized=hypothesized_population_mean_iq,
    population_std_dev=known_population_std_dev_iq,
    sample_size=sample_size_iq,
    alpha=significance_level,
    test_type='right-tailed'
)
print(f"Calculated Z-statistic: {z_stat_right:.4f}")
print(f"P-value (right-tailed): {p_val_right:.4f}")
print(f"Decision: {decision_result_right}")
'''
```

```python
#Q16.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for the mean of a dataset.
    Assumes a normal distribution and uses the t-distribution for small sample sizes
    or when population standard deviation is unknown.

    Args:
        data (list or np.array): The input dataset.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).

    Returns:
        tuple: A tuple containing the sample mean, margin of error, and the confidence interval (lower, upper).
    """
    data_np = np.array(data)
    n = len(data_np)

    if n < 2:
        raise ValueError("Dataset must contain at least two data points to calculate a confidence interval.")

    sample_mean = np.mean(data_np)
    sample_std_dev = np.std(data_np, ddof=1) # Use ddof=1 for sample standard deviation (unbiased estimator)

    # For confidence intervals of the mean, we use the t-distribution
    # especially when the population standard deviation is unknown (which is common).
    # The t-distribution is more appropriate for smaller sample sizes.
    degrees_freedom = n - 1

    # Calculate the t-critical value
    # For a two-tailed interval, we need alpha/2 in each tail
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha/2, degrees_freedom)

    # Calculate the margin of error
    margin_of_error = t_critical * (sample_std_dev / np.sqrt(n))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, margin_of_error, (lower_bound, upper_bound)

# --- Example Usage ---
# Dataset of exam scores
exam_scores = [78, 85, 90, 72, 88, 92, 75, 80, 83, 95, 68, 70, 81, 86, 79]
confidence_lvl = 0.95

print(f"Dataset: {exam_scores}")
print(f"Desired Confidence Level: {confidence_lvl * 100}%\n")

mean, moe, ci = calculate_confidence_interval(exam_scores, confidence_lvl)

print(f"Sample Mean: {mean:.2f}")
print(f"Margin of Error: {moe:.2f}")
print(f"Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

# Interpretation
print("\n--- Interpretation ---")
print(f"We are {confidence_lvl * 100}% confident that the true population mean")
print(f"of exam scores lies within the interval ({ci[0]:.2f}, {ci[1]:.2f}).")
print("This means if we were to repeat this sampling process many times,")
print(f"{confidence_lvl * 100}% of the calculated confidence intervals would contain the true population mean.")

# Example with a different confidence level
confidence_lvl_99 = 0.99
mean_99, moe_99, ci_99 = calculate_confidence_interval(exam_scores, confidence_lvl_99)
print(f"\nFor a {confidence_lvl_99 * 100}% confidence level:")
print(f"Confidence Interval: ({ci_99[0]:.2f}, {ci_99[1]:.2f})")
print("Notice that a higher confidence level results in a wider interval,")
print("reflecting greater certainty but less precision.")
```

```python
#Q19.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf_and_interpret(lambda_param, max_k=10):
    """
    Calculates and interprets the Cumulative Distribution Function (CDF)
    of a Poisson distribution.

    Args:
        lambda_param (float): The average rate of events (λ).
        max_k (int): The maximum integer value to calculate CDF for plotting.
    """
    if lambda_param <= 0:
        raise ValueError("Lambda parameter must be positive.")
    if max_k <= 0:
        raise ValueError("max_k must be positive.")

    print(f"--- Poisson CDF Calculation and Interpretation (λ = {lambda_param}) ---")

    # Generate an array of possible event counts (k)
    k_values = np.arange(0, max_k + 1)

    # Calculate the CDF for each k value
    cdf_values = poisson.cdf(k_values, mu=lambda_param)

    print("\nCumulative Probabilities (CDF):")
    print("k (Events) | P(X <= k)")
    print("-----------|----------")
    for k, cdf_val in zip(k_values, cdf_values):
        print(f" {k:<10} | {cdf_val:.4f}")

    # Interpretation
    print("\n--- Interpretation Examples ---")

    # Example 1: Probability of at most 3 events
    target_k_1 = 3
    if target_k_1 <= max_k:
        prob_at_most_3 = poisson.cdf(target_k_1, mu=lambda_param)
        print(f"P(X <= {target_k_1}) = {prob_at_most_3:.4f}")
        print(f"This means there is a {prob_at_most_3*100:.2f}% chance of observing {target_k_1} or fewer events.")
    else:
        print(f"Cannot interpret for P(X <= {target_k_1}) as it's outside the calculated range (0 to {max_k}).")

    # Example 2: Probability of at least 5 events
    target_k_2 = 5
    if target_k_2 - 1 >= 0 and target_k_2 - 1 <= max_k:
        prob_at_least_5 = 1 - poisson.cdf(target_k_2 - 1, mu=lambda_param)
        print(f"P(X >= {target_k_2}) = 1 - P(X <= {target_k_2-1}) = {prob_at_least_5:.4f}")
        print(f"This means there is a {prob_at_least_5*100:.2f}% chance of observing {target_k_2} or more events.")
    else:
         print(f"Cannot interpret for P(X >= {target_k_2}) directly from this range.")

    # Example 3: Probability of exactly a certain number of events (using PMF)
    target_k_exact = int(lambda_param) # Often, the most probable value is around lambda
    prob_exact = poisson.pmf(target_k_exact, mu=lambda_param)
    print(f"P(X = {target_k_exact}) = {prob_exact:.4f}")
    print(f"This is the probability of observing exactly {target_k_exact} events.")


    # Plotting the CDF
    plt.figure(figsize=(10, 6))
    plt.step(k_values, cdf_values, where='post', color='darkviolet', linestyle='-', marker='o', markersize=6)
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (λ = {lambda_param})')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('P(X <= k)')
    plt.xticks(k_values)
    plt.yticks(np.linspace(0, 1, 11))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# --- Example Usage ---
poisson_cdf_and_interpret(lambda_param=2.5, max_k=10)

print("\n" + "="*50 + "\n")
poisson_cdf_and_interpret(lambda_param=5, max_k=15)
'''
```

```python
#Q23.Implement Z-score normalization on a dataset and visualize its transformation.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm # For plotting theoretical normal distribution

def z_score_normalize_and_visualize(data, original_name="Original Data"):
    """
    Performs Z-score normalization (standardization) on a dataset
    and visualizes the original and normalized distributions.

    Z-score normalization transforms data to have a mean of 0 and a standard deviation of 1.
    Formula: Z = (X - μ) / σ

    Args:
        data (list or np.array): The input dataset.
        original_name (str): A name for the original dataset for plotting.

    Returns:
        np.array: The Z-score normalized dataset.
    """
    data_np = np.array(data)

    if len(data_np) == 0:
        print("Error: Dataset is empty. Cannot normalize.")
        return np.array([])
    if len(data_np) == 1:
        print("Warning: Dataset has only one element. Z-score will be 0.")
        return np.array([0.0])

    # Calculate mean and standard deviation of the original data
    original_mean = np.mean(data_np)
    original_std = np.std(data_np)

    # Perform Z-score normalization
    if original_std == 0:
        # If std dev is 0, all values are the same, Z-scores will be 0
        normalized_data = np.zeros_like(data_np, dtype=float)
        print("Warning: Standard deviation of original data is zero. Normalized data will be all zeros.")
    else:
        normalized_data = (data_np - original_mean) / original_std

    # Calculate mean and standard deviation of the normalized data
    normalized_mean = np.mean(normalized_data)
    normalized_std = np.std(normalized_data)

    print(f"--- Z-score Normalization for '{original_name}' ---")
    print(f"Original Data Mean: {original_mean:.2f}")
    print(f"Original Data Std Dev: {original_std:.2f}")
    print(f"Normalized Data Mean: {normalized_mean:.4f} (should be close to 0)")
    print(f"Normalized Data Std Dev: {normalized_std:.4f} (should be close to 1)\n")

    # --- Visualization ---
    plt.figure(figsize=(12, 6))

    # Plot histogram of original data
    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    plt.hist(data_np, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black', label=f'{original_name} (μ={original_mean:.2f}, σ={original_std:.2f})')
    plt.title(f'Original Distribution of {original_name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency/Probability Density')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()

    # Plot histogram of normalized data
    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    plt.hist(normalized_data, bins=30, density=True, alpha=0.7, color='lightcoral', edgecolor='black', label='Normalized Data')
    # Overlay theoretical standard normal PDF for comparison
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, 0, 1) # Standard normal distribution (mean=0, std=1)
    plt.plot(x, p, 'blue', linewidth=2, label='Standard Normal PDF')

    plt.title('Z-score Normalized Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency/Probability Density')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()

    plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
    plt.show()

    return normalized_data

# --- Example Usage ---
# Example 1: Data from a normal distribution
data_normal = np.random.normal(loc=100, scale=15, size=1000)
normalized_normal_data = z_score_normalize_and_visualize(data_normal, "Exam Scores")

print("\n" + "="*50 + "\n")

# Example 2: Data from a skewed distribution (e.g., exponential)
data_skewed = np.random.exponential(scale=5, size=1000)
normalized_skewed_data = z_score_normalize_and_visualize(data_skewed, "Waiting Times")
'''
```

```python
#Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
'''
import numpy as np
from scipy.stats import norm

def z_test_one_sample(sample_mean, population_mean, population_std, sample_size, alpha):
    """
    Performs a one-sample Z-test.

    H0: sample_mean = population_mean (Null Hypothesis)
    Ha: sample_mean != population_mean (Alternative Hypothesis - two-tailed)

    Args:
        sample_mean (float): The mean of the sample.
        population_mean (float): The known mean of the population under the null hypothesis.
        population_std (float): The known standard deviation of the population.
        sample_size (int): The number of observations in the sample.
        alpha (float): The significance level (e.g., 0.05).

    Returns:
        tuple: (z_statistic, p_value, decision)
    """
    if sample_size <= 0:
        raise ValueError("Sample size must be greater than 0.")
    if population_std <= 0:
        raise ValueError("Population standard deviation must be greater than 0.")

    # Calculate the standard error of the mean
    std_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / std_error

    # Calculate the two-tailed P-value
    # norm.cdf(z_statistic) gives P(Z <= z_statistic)
    # For two-tailed, P = 2 * min(P(Z <= z), P(Z >= z))
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Make a decision
    if p_value < alpha:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    return z_statistic, p_value, decision

# --- Example Usage ---
# Scenario: A school claims its students' average IQ is 100 with a standard deviation of 15.
# A sample of 30 students has an average IQ of 105.
# We want to know if this sample mean is significantly different from the population mean.

sample_mean_iq = 105
population_mean_iq = 100
population_std_iq = 15
sample_size_iq = 30
alpha_iq = 0.05 # 5% significance level

z_stat_iq, p_val_iq, decision_iq = z_test_one_sample(
    sample_mean_iq, population_mean_iq, population_std_iq, sample_size_iq, alpha_iq
)

print("--- One-Sample Z-Test Results ---")
print(f"Sample Mean: {sample_mean_iq}")
print(f"Hypothesized Population Mean (H0): {population_mean_iq}")
print(f"Known Population Standard Deviation: {population_std_iq}")
print(f"Sample Size: {sample_size_iq}")
print(f"Significance Level (alpha): {alpha_iq}\n")

print(f"Calculated Z-statistic: {z_stat_iq:.4f}")
print(f"P-value: {p_val_iq:.4f}")
print(f"Decision: {decision_iq}")

# --- Interpretation ---
print("\n--- Interpretation ---")
if decision_iq == "Reject the null hypothesis":
    print(f"Since the P-value ({p_val_iq:.4f}) is less than the significance level ({alpha_iq}),")
    print("we have sufficient evidence to reject the null hypothesis.")
    print(f"This suggests that the sample mean IQ ({sample_mean_iq}) is significantly different")
    print(f"from the hypothesized population mean IQ ({population_mean_iq}).")
else:
    print(f"Since the P-value ({p_val_iq:.4f}) is greater than or equal to the significance level ({alpha_iq}),")
    print("we do not have sufficient evidence to reject the null hypothesis.")
    print(f"This suggests that the sample mean IQ ({sample_mean_iq}) is not significantly different")
    print(f"from the hypothesized population mean IQ ({population_mean_iq}).")
'''
```

```python
#Q2.Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
'''
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def simulate_and_test_mean(population_mean, population_std, sample_size, num_simulations, alpha):
    """
    Simulates random samples, calculates sample means, and performs a Z-test
    to demonstrate hypothesis testing and P-value calculation.

    Args:
        population_mean (float): The true mean of the population.
        population_std (float): The true standard deviation of the population.
        sample_size (int): The size of each sample to draw.
        num_simulations (int): The number of samples to simulate.
        alpha (float): The significance level.
    """
    sample_means = []
    p_values = []
    reject_count = 0

    print(f"--- Simulating {num_simulations} Samples for Z-Test ---")
    print(f"Population Mean (μ): {population_mean}")
    print(f"Population Std Dev (σ): {population_std}")
    print(f"Sample Size (n): {sample_size}")
    print(f"Significance Level (α): {alpha}\n")

    for i in range(num_simulations):
        # Simulate a sample from the population
        sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
        current_sample_mean = np.mean(sample)
        sample_means.append(current_sample_mean)

        # Calculate Z-statistic for this sample
        std_error = population_std / np.sqrt(sample_size)
        z_statistic = (current_sample_mean - population_mean) / std_error

        # Calculate two-tailed P-value
        p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
        p_values.append(p_value)

        # Check if null hypothesis is rejected for this sample
        if p_value < alpha:
            reject_count += 1

    rejection_rate = reject_count / num_simulations

    print(f"Average Sample Mean over {num_simulations} simulations: {np.mean(sample_means):.2f}")
    print(f"Average P-value: {np.mean(p_values):.4f}")
    print(f"Number of times H0 was rejected (P-value < {alpha}): {reject_count} out of {num_simulations}")
    print(f"Empirical Rejection Rate: {rejection_rate:.4f} (should be close to alpha if H0 is true)\n")

    # --- Visualization of Sample Means Distribution ---
    plt.figure(figsize=(10, 6))
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='lightgreen', edgecolor='black',
             label='Distribution of Sample Means')

    # Overlay theoretical sampling distribution of the mean
    x_mean = np.linspace(min(sample_means), max(sample_means), 100)
    theoretical_std_error = population_std / np.sqrt(sample_size)
    pdf_means = norm.pdf(x_mean, population_mean, theoretical_std_error)
    plt.plot(x_mean, pdf_means, 'r--', linewidth=2, label='Theoretical Sampling Distribution')

    plt.axvline(population_mean, color='blue', linestyle=':', label='Population Mean (μ)')

    plt.title(f'Distribution of Sample Means (n={sample_size}, {num_simulations} Simulations)')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

# --- Example Usage ---
# Assume a population with mean 100 and standard deviation 10.
# We are testing if a sample mean is significantly different from 100.
simulate_and_test_mean(
    population_mean=100,
    population_std=10,
    sample_size=30,
    num_simulations=5000, # Perform many simulations to see the trend
    alpha=0.05)
'''
```

```python
#Q3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
'''
import numpy as np
from scipy.stats import norm

def perform_one_sample_z_test(sample_data, population_mean_h0, population_std_known, alpha_level):
    """
    Performs a one-sample Z-test to compare a sample mean to a known population mean.

    Null Hypothesis (H0): The sample mean is equal to the population mean (μ_sample = μ_population)
    Alternative Hypothesis (Ha): The sample mean is not equal to the population mean (μ_sample != μ_population)
                                 (This is a two-tailed test by default)

    Args:
        sample_data (array-like): The observed sample data.
        population_mean_h0 (float): The hypothesized population mean under the null hypothesis.
        population_std_known (float): The known standard deviation of the population.
        alpha_level (float): The significance level (e.g., 0.05).

    Returns:
        dict: A dictionary containing test results (sample_mean, z_statistic, p_value, decision).
    """
    sample_np = np.array(sample_data)
    n = len(sample_np)

    if n == 0:
        return {"error": "Sample data is empty."}
    if population_std_known <= 0:
        return {"error": "Population standard deviation must be positive."}

    sample_mean = np.mean(sample_np)
    standard_error = population_std_known / np.sqrt(n)

    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean_h0) / standard_error

    # Calculate P-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Make decision
    if p_value < alpha_level:
        decision = "Reject H0"
        conclusion = f"There is sufficient evidence to conclude that the sample mean ({sample_mean:.2f}) is significantly different from the hypothesized population mean ({population_mean_h0})."
    else:
        decision = "Fail to Reject H0"
        conclusion = f"There is not sufficient evidence to conclude that the sample mean ({sample_mean:.2f}) is significantly different from the hypothesized population mean ({population_mean_h0})."

    results = {
        "sample_mean": sample_mean,
        "hypothesized_population_mean": population_mean_h0,
        "population_std_known": population_std_known,
        "sample_size": n,
        "significance_level": alpha_level,
        "z_statistic": z_statistic,
        "p_value": p_value,
        "decision": decision,
        "conclusion": conclusion
    }
    return results

# --- Example Usage ---
# Scenario: A soft drink company claims that its soda cans contain an average of 330 ml.
# The volume is known to have a standard deviation of 5 ml.
# A quality control inspector takes a sample of 40 cans and finds the average volume to be 328 ml.
# Is there enough evidence to say the average volume is different from 330 ml at 5% significance?

# Simulate sample data for demonstration (in a real scenario, you'd have actual measurements)
np.random.seed(42) # for reproducibility
sample_data_cans = np.random.normal(loc=328, scale=5, size=40) # Sample with mean 328

hypothesized_mean = 330
known_std = 5
significance_level = 0.05

test_results = perform_one_sample_z_test(
    sample_data_cans,
    hypothesized_mean,
    known_std,
    significance_level
)

print("--- One-Sample Z-Test for Soda Can Volume ---")
for key, value in test_results.items():
    if isinstance(value, float):
        print(f"{key.replace('_', ' ').capitalize()}: {value:.4f}")
    else:
        print(f"{key.replace('_', ' ').capitalize()}: {value}")

'''
```

```python
#Q4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
'''
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def plot_two_tailed_z_test(z_statistic, alpha, plot_range=(-4, 4)):
    """
    Visualizes the decision regions for a two-tailed Z-test on a standard normal distribution.

    Args:
        z_statistic (float): The calculated Z-statistic.
        alpha (float): The significance level.
        plot_range (tuple): The min and max values for the x-axis of the plot.
    """
    # Create a range of x-values for the standard normal distribution
    x = np.linspace(plot_range[0], plot_range[1], 500)
    # Calculate the PDF for the standard normal distribution
    pdf_values = norm.pdf(x, 0, 1)

    # Calculate the critical Z-values for the two-tailed test
    # alpha/2 for each tail
    critical_z_lower = norm.ppf(alpha / 2)
    critical_z_upper = norm.ppf(1 - alpha / 2)

    # Calculate the P-value for the calculated Z-statistic
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Determine the decision
    if p_value < alpha:
        decision = "Reject H0"
        decision_color = 'red'
    else:
        decision = "Fail to Reject H0"
        decision_color = 'green'

    print(f"--- Two-Tailed Z-Test Visualization ---")
    print(f"Calculated Z-statistic: {z_statistic:.4f}")
    print(f"Significance Level (α): {alpha}")
    print(f"Critical Z-values: ({critical_z_lower:.4f}, {critical_z_upper:.4f})")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}\n")

    # --- Plotting ---
    plt.figure(figsize=(12, 7))
    plt.plot(x, pdf_values, color='blue', lw=2, label='Standard Normal PDF (μ=0, σ=1)')

    # Shade the critical regions (rejection regions)
    x_reject_lower = x[x <= critical_z_lower]
    x_reject_upper = x[x >= critical_z_upper]
    plt.fill_between(x_reject_lower, 0, norm.pdf(x_reject_lower, 0, 1), color='red', alpha=0.3, label='Rejection Region (α/2)')
    plt.fill_between(x_reject_upper, 0, norm.pdf(x_reject_upper, 0, 1), color='red', alpha=0.3)

    # Plot the calculated Z-statistic
    plt.axvline(z_statistic, color=decision_color, linestyle='--', lw=2,
                label=f'Calculated Z: {z_statistic:.2f} ({decision})')

    # Add critical Z-value lines
    plt.axvline(critical_z_lower, color='red', linestyle=':', lw=1.5, label=f'Critical Z: ±{abs(critical_z_lower):.2f}')
    plt.axvline(critical_z_upper, color='red', linestyle=':', lw=1.5)

    # Add labels and title
    plt.title(f'Two-Tailed Z-Test: Decision Region (α={alpha})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

# --- Example Usage ---
# Example 1: Z-statistic falls in the rejection region
calculated_z1 = 2.15
alpha1 = 0.05
plot_two_tailed_z_test(calculated_z1, alpha1)

# Example 2: Z-statistic falls in the fail-to-reject region
calculated_z2 = 0.8
alpha2 = 0.05
plot_two_tailed_z_test(calculated_z2, alpha2)

# Example 3: Higher significance level
calculated_z3 = 1.9
alpha3 = 0.10
plot_two_tailed_z_test(calculated_z3, alpha3)
'''
```

```python
#Q5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
'''
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def visualize_type_errors(pop_mean_h0, pop_std, sample_size, alpha, true_pop_mean_ha=None):
    """
    Calculates and visualizes Type I and Type II errors for a one-sample Z-test.

    Type I Error (α): Rejecting a true null hypothesis. (False Positive)
    Type II Error (β): Failing to reject a false null hypothesis. (False Negative)
    Power (1 - β): Correctly rejecting a false null hypothesis.

    Args:
        pop_mean_h0 (float): Hypothesized population mean under the Null Hypothesis (H0).
        pop_std (float): Known population standard deviation.
        sample_size (int): Size of the sample.
        alpha (float): Significance level (probability of Type I error).
        true_pop_mean_ha (float, optional): The true population mean under the Alternative Hypothesis (Ha).
                                            If None, only Type I error is visualized.
    """
    # Calculate critical Z-values for a two-tailed test
    # This determines the rejection regions under H0
    critical_z_lower = norm.ppf(alpha / 2)
    critical_z_upper = norm.ppf(1 - alpha / 2)

    # Convert critical Z-values to sample means (critical values for sample mean)
    std_error = pop_std / np.sqrt(sample_size)
    critical_sample_mean_lower = pop_mean_h0 + critical_z_lower * std_error
    critical_sample_mean_upper = pop_mean_h0 + critical_z_upper * std_error

    print(f"--- Visualization of Type I and Type II Errors ---")
    print(f"Null Hypothesis (H0) Mean: {pop_mean_h0}")
    print(f"Population Std Dev (σ): {pop_std}")
    print(f"Sample Size (n): {sample_size}")
    print(f"Significance Level (α) for Type I Error: {alpha}")
    print(f"Critical Sample Means: ({critical_sample_mean_lower:.2f}, {critical_sample_mean_upper:.2f})\n")

    # --- Plotting ---
    plt.figure(figsize=(12, 7))
    x_range = np.linspace(pop_mean_h0 - 4 * std_error, pop_mean_h0 + 4 * std_error, 500)

    # Plot the distribution under the Null Hypothesis (H0)
    pdf_h0 = norm.pdf(x_range, loc=pop_mean_h0, scale=std_error)
    plt.plot(x_range, pdf_h0, color='blue', lw=2, label='Distribution under H0 (True μ = H0 Mean)')
    plt.axvline(pop_mean_h0, color='blue', linestyle=':', label=f'H0 Mean: {pop_mean_h0}')

    # Shade Type I Error (Alpha) regions
    x_type1_lower = x_range[x_range <= critical_sample_mean_lower]
    x_type1_upper = x_range[x_range >= critical_sample_mean_upper]
    plt.fill_between(x_type1_lower, 0, norm.pdf(x_type1_lower, pop_mean_h0, std_error),
                     color='red', alpha=0.3, label='Type I Error (α/2)')
    plt.fill_between(x_type1_upper, 0, norm.pdf(x_type1_upper, pop_mean_h0, std_error),
                     color='red', alpha=0.3)

    # Add critical lines
    plt.axvline(critical_sample_mean_lower, color='red', linestyle='--', lw=1.5, label='Critical Values')
    plt.axvline(critical_sample_mean_upper, color='red', linestyle='--', lw=1.5)

    # If true_pop_mean_ha is provided, visualize Type II Error (Beta) and Power
    if true_pop_mean_ha is not None:
        # Calculate the distribution under the Alternative Hypothesis (Ha)
        # Assuming Ha is a specific mean difference
        pdf_ha = norm.pdf(x_range, loc=true_pop_mean_ha, scale=std_error)
        plt.plot(x_range, pdf_ha, color='green', lw=2, label=f'Distribution under Ha (True μ = {true_pop_mean_ha})')
        plt.axvline(true_pop_mean_ha, color='green', linestyle=':', label=f'Ha Mean: {true_pop_mean_ha}')

        # Calculate Beta (Type II Error)
        # This is the area under Ha's curve that falls into the "fail to reject H0" region
        beta = norm.cdf(critical_sample_mean_upper, loc=true_pop_mean_ha, scale=std_error) - \
               norm.cdf(critical_sample_mean_lower, loc=true_pop_mean_ha, scale=std_error)
        power = 1 - beta

        x_type2_region = x_range[(x_range > critical_sample_mean_lower) & (x_range < critical_sample_mean_upper)]
        plt.fill_between(x_type2_region, 0, norm.pdf(x_type2_region, true_pop_mean_ha, std_error),
                         color='orange', alpha=0.4, label='Type II Error (β)')

        print(f"True Population Mean under Ha: {true_pop_mean_ha}")
        print(f"Probability of Type II Error (β): {beta:.4f}")
        print(f"Power (1 - β): {power:.4f}\n")

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

# --- Example Usage ---
# Scenario: Testing if a new drug increases response time from 100ms.
# Known population std dev = 10ms. Sample size = 30. Alpha = 0.05.

# Example 1: Visualize only Type I error (assuming H0 is true)
visualize_type_errors(
    pop_mean_h0=100,
    pop_std=10,
    sample_size=30,
    alpha=0.05
)

# Example 2: Visualize Type I and Type II errors (assuming Ha is true with a specific mean)
# True mean under Ha is 105 (drug actually increases response time)
visualize_type_errors(
    pop_mean_h0=100,
    pop_std=10,
    sample_size=30,
    alpha=0.05,
    true_pop_mean_ha=105
)

# Example 3: With smaller sample size (increases std error, higher Beta)
visualize_type_errors(
    pop_mean_h0=100,
    pop_std=10,
    sample_size=10, # Smaller sample size
    alpha=0.05,
    true_pop_mean_ha=105
)

# Example 4: With larger alpha (decreases Beta, increases Type I)
visualize_type_errors(
    pop_mean_h0=100,
    pop_std=10,
    sample_size=30,
    alpha=0.10, # Larger alpha
    true_pop_mean_ha=105)

'''
```

```python
#Q6.Write a Python program to perform an independent T-test and interpret the results.
'''
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def perform_independent_ttest(group1, group2, alpha=0.05):
    """
    Perform an independent t-test between two groups and interpret the results.

    Parameters:
    - group1: First sample data (array-like)
    - group2: Second sample data (array-like)
    - alpha: Significance level (default 0.05)
    """

    # Basic descriptive statistics
    n1, n2 = len(group1), len(group2)
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)

    # Perform the t-test
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)

    # Calculate degrees of freedom
    df = n1 + n2 - 2

    # Calculate 95% confidence interval for the difference
    pooled_std = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / df)
    margin_error = stats.t.ppf(1 - alpha/2, df) * pooled_std * np.sqrt(1/n1 + 1/n2)
    diff = mean1 - mean2
    ci_lower = diff - margin_error
    ci_upper = diff + margin_error

    # Print results
    print("\n=== Independent T-Test Results ===")
    print(f"Group 1: n = {n1}, Mean = {mean1:.2f}, SD = {std1:.2f}")
    print(f"Group 2: n = {n2}, Mean = {mean2:.2f}, SD = {std2:.2f}")
    print(f"\nT-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of freedom: {df}")
    print(f"95% CI for difference: ({ci_lower:.2f}, {ci_upper:.2f})")

    # Interpret results
    print("\n=== Interpretation ===")
    if p_value < alpha:
        print(f"Reject the null hypothesis (p < {alpha}).")
        if diff > 0:
            print("Group 1 mean is significantly greater than Group 2.")
        else:
            print("Group 2 mean is significantly greater than Group 1.")
    else:
        print(f"Fail to reject the null hypothesis (p ≥ {alpha}).")
        print("No significant difference between groups detected.")

    # Plot the data
    plt.figure(figsize=(10, 6))
    plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
    plt.title('Distribution Comparison')
    plt.ylabel('Values')
    plt.grid(True)
    plt.show()

    return t_stat, p_value

# Example usage with simulated data
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)  # Mean=50, SD=10
group2 = np.random.normal(55, 10, 30)  # Mean=55, SD=10

# Perform the t-test
t_stat, p_value = perform_independent_ttest(group1, group2)
'''
```

```python
#Q7.Perform a paired sample T-test using Python and visualize the comparison results.
'''
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

def paired_t_test_and_visualize(before_data, after_data, alpha, data_name="Measure"):
    """
    Performs a paired samples T-test and visualizes the distribution of differences.

    A paired t-test is used when the samples are dependent, typically when
    the same subjects are measured twice (e.g., "before" and "after" an intervention).
    H0: μ_diff = 0 (Mean difference is zero)
    Ha: μ_diff ≠ 0 (Mean difference is not zero - two-tailed)

    Args:
        before_data (array-like): Data collected before the intervention.
        after_data (array-like): Data collected after the intervention.
        alpha (float): The significance level.
        data_name (str): Name of the measure for plot titles.

    Returns:
        dict: Test results.
    """
    before_np = np.array(before_data)
    after_np = np.array(after_data)

    if len(before_np) != len(after_np):
        return {"error": "Before and After datasets must have the same number of observations for a paired t-test."}
    if len(before_np) < 2:
        return {"error": "Paired samples must have at least 2 pairs for a T-test."}

    # Calculate the differences
    differences = after_np - before_np

    # Perform the paired samples t-test
    # ttest_rel returns (t-statistic, p-value)
    t_statistic, p_value = ttest_rel(before_np, after_np) # Note: order matters for t-stat sign

    # Make decision
    if p_value < alpha:
        decision = "Reject H0"
        conclusion = f"There is sufficient evidence to conclude a significant change in {data_name} after the intervention."
    else:
        decision = "Fail to Reject H0"
        conclusion = f"There is not sufficient evidence to conclude a significant change in {data_name} after the intervention."

    results = {
        "before_mean": np.mean(before_np),
        "after_mean": np.mean(after_np),
        "mean_difference": np.mean(differences),
        "std_difference": np.std(differences, ddof=1),
        "sample_size": len(before_np),
        "t_statistic": t_statistic,
        "p_value": p_value,
        "significance_level": alpha,
        "decision": decision,
        "conclusion": conclusion
    }
    print(f"--- Paired Samples T-Test: {data_name} Comparison ---")
    for key, value in results.items():
        if isinstance(value, float):
            print(f"{key.replace('_', ' ').capitalize()}: {value:.4f}")
        else:
            print(f"{key.replace('_', ' ').capitalize()}: {value}")

    # --- Visualization ---
    plt.figure(figsize=(12, 6))

    # Plot 1: Box plot of Before and After
    plt.subplot(1, 2, 1)
    plt.boxplot([before_np, after_np], labels=['Before', 'After'])
    plt.title(f'Comparison of {data_name} (Before vs. After)')
    plt.ylabel(data_name)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Plot 2: Histogram of Differences
    plt.subplot(1, 2, 2)
    plt.hist(differences, bins=10, edgecolor='black', alpha=0.7, color='lightseagreen')
    plt.axvline(np.mean(differences), color='red', linestyle='--', label=f'Mean Difference: {np.mean(differences):.2f}')
    plt.axvline(0, color='grey', linestyle=':', label='Zero Difference')
    plt.title(f'Distribution of Differences in {data_name}')
    plt.xlabel('Difference (After - Before)')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()

    plt.tight_layout()
    plt.show()

    return results

# --- Example Usage ---
# Scenario 1: A new training program's effect on reaction time.
# Lower reaction time is better.
np.random.seed(42) # for reproducibility
before_rt = np.random.normal(loc=250, scale=20, size=20) # Reaction times before training
after_rt = np.random.normal(loc=230, scale=18, size=20)  # Reaction times after training

paired_t_test_and_visualize(before_rt, after_rt, alpha=0.05, data_name="Reaction Time (ms)")

print("\n" + "="*70 + "\n")

# Scenario 2: Effect of a diet on weight.
# Higher weight is worse.
before_weight = np.random.normal(loc=75, scale=5, size=25)
after_weight = np.random.normal(loc=74, scale=4, size=25)

paired_t_test_and_visualize(before_weight, after_weight, alpha=0.01, data_name="Weight (kg)")
'''
```

```python
#Q14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
'''
import numpy as np
from scipy.stats import chisquare

def chi_square_goodness_of_fit(observed_frequencies, expected_frequencies, alpha):
    """
    Performs a Chi-square goodness-of-fit test.

    This test determines if an observed frequency distribution differs significantly
    from an expected frequency distribution.

    H0: The observed data fit the expected distribution (no significant difference).
    Ha: The observed data do not fit the expected distribution (significant difference).

    Args:
        observed_frequencies (array-like): The observed counts for each category.
        expected_frequencies (array-like): The expected counts for each category
                                            under the null hypothesis. Must have the same length
                                            as observed_frequencies.
        alpha (float): The significance level (e.0.05).

    Returns:
        dict: A dictionary containing test results (chi2_statistic, p_value, decision).
    """
    observed_np = np.array(observed_frequencies)
    expected_np = np.array(expected_frequencies)

    if not (len(observed_np) == len(expected_np) and len(observed_np) > 0):
        return {"error": "Observed and expected frequencies must be non-empty and have the same number of categories."}
    if np.any(observed_np < 0) or np.any(expected_np < 0):
        return {"error": "Frequencies cannot be negative."}
    if np.any(expected_np == 0):
        # Chi-square test is sensitive to expected frequencies of 0,
        # typically categories with <5 expected counts are problematic and might need merging.
        print("Warning: Some expected frequencies are zero. Chi-square test assumptions might be violated.")

    # Perform the Chi-square goodness-of-fit test
    # chisquare returns: chi2_statistic, p_value
    # Degrees of freedom = (number of categories) - 1 (if expected proportions are pre-specified)
    # If parameters of the expected distribution are estimated from the observed data,
    # then df = (number of categories) - 1 - (number of parameters estimated).
    # Here, we assume expected frequencies are fully pre-specified.
    chi2_statistic, p_value = chisquare(f_obs=observed_np, f_exp=expected_np)
    degrees_of_freedom = len(observed_np) - 1

    # Make decision
    if p_value < alpha:
        decision = "Reject H0"
        conclusion = f"There is sufficient evidence to conclude that the observed distribution does not fit the expected distribution."
    else:
        decision = "Fail to Reject H0"
        conclusion = f"There is not sufficient evidence to conclude that the observed distribution does not fit the expected distribution. It appears to fit."

    results = {
        "observed_frequencies": observed_np.tolist(),
        "expected_frequencies": expected_np.tolist(),
        "chi2_statistic": chi2_statistic,
        "p_value": p_value,
        "degrees_of_freedom": degrees_of_freedom,
        "significance_level": alpha,
        "decision": decision,
        "conclusion": conclusion
    }
    return results

# --- Example Usage ---
# Scenario 1: M&M's color distribution
# Claimed distribution by M&M's: Blue (24%), Orange (20%), Green (16%), Yellow (14%), Red (13%), Brown (13%)
# Observed counts from a bag of 100 M&M's:

total_observed = 100
observed_colors = np.array([22, 23, 14, 15, 12, 14]) # Blue, Orange, Green, Yellow, Red, Brown

# Calculate expected counts based on total observed and claimed proportions
expected_proportions = np.array([0.24, 0.20, 0.16, 0.14, 0.13, 0.13])
expected_colors = expected_proportions * total_observed

alpha = 0.05

test_results_mm = chi_square_goodness_of_fit(observed_colors, expected_colors, alpha)

print("--- Chi-Square Goodness-of-Fit Test: M&M's Colors ---")
print(f"Observed Frequencies: {test_results_mm['observed_frequencies']}")
print(f"Expected Frequencies: {[round(x, 2) for x in test_results_mm['expected_frequencies']]}") # Round for display consistency
print(f"Chi2 Statistic: {test_results_mm['chi2_statistic']:.4f}")
print(f"P-value: {test_results_mm['p_value']:.4f}")
print(f"Degrees of Freedom: {test_results_mm['degrees_of_freedom']}")
print(f"Significance Level (alpha): {test_results_mm['significance_level']}\n")
print(f"Decision: {test_results_mm['decision']}")
print(f"Conclusion: {test_results_mm['conclusion']}")

print("\n" + "="*70 + "\n")

# Scenario 2: Dice roll fairness
# Expected: each face (1-6) appears equally often in 60 rolls (10 times each)
observed_rolls = np.array([8, 12, 9, 11, 10, 10]) # Observed counts for faces 1 to 6
expected_rolls = np.array([10, 10, 10, 10, 10, 10])

test_results_dice = chi_square_goodness_of_fit(observed_rolls, expected_rolls, 0.05)
print("--- Chi-Square Goodness-of-Fit Test: Dice Rolls ---")
print(f"Observed Frequencies: {test_results_dice['observed_frequencies']}")
print(f"Expected Frequencies: {test_results_dice['expected_frequencies']}")
print(f"Chi2 Statistic: {test_results_dice['chi2_statistic']:.4f}")
print(f"P-value: {test_results_dice['p_value']:.4f}")
print(f"Degrees of Freedom: {test_results_dice['degrees_of_freedom']}")
print(f"Significance Level (alpha): {test_results_dice['significance_level']}\n")
print(f"Decision: {test_results_dice['decision']}")
print(f"Conclusion: {test_results_dice['conclusion']}")
'''
```
